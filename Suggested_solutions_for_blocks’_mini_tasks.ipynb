```python
# Download stopwords (only needed once)
nltk.download('stopwords')

# Create a list of English stopwords
stop_words = set(stopwords.words('english'))

# Initialize the stemmer
stemmer = PorterStemmer()

# Define the text preprocessing function
def preprocess_text(text):
    """
    This function:
    1. Converts text to lowercase
    2. Removes punctuation
    3. Removes stopwords
    4. Converts words to their stems
    """

    # Check for missing values
    if pd.isna(text):
        return ""

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)

    # Split text into words
    words = text.split()

    # Remove stopwords and apply stemming
    processed_words = [
        stemmer.stem(word)
        for word in words
        if word not in stop_words
    ]

    # Join words back into a single string
    return " ".join(processed_words)
```

```python
# Apply the function to the full_description column
# This may take 2-3 minutes...
df["clean_description"] = df["full_description"].apply(preprocess_text)

# Display results
df[["full_description", "clean_description"]].head()
```

```python
# It is important to apply the same function preprocess_text() to our AI keywords
ai_keywords_clean = [preprocess_text(keyword) for keyword in ai_keywords]

# Display results
ai_keywords_clean
```

```python
# Create a new column for AI-related trademarks
df['AI'] = 0

# Check whether any AI keyword appears in the cleaned description
# A lambda function (below) in Python is a short and simple function written in a single line.
# It is often used when you need a quick function for a short task and do not want to define a full function using def.
for keyword in ai_keywords_clean:
    df['AI'] = df.apply(
        lambda row: 1 if keyword in row['clean_description'] else row['AI'],
        axis=1
    )

# Display the count of AI-related trademarks
print(f"Number of AI-related trademarks: {df['AI'].sum()}")

# Display the first few rows with the new 'AI' column
df[['full_description', 'clean_description', 'AI']].head()
```

```python
# Display 3 random examples where AI == 1

print(
    df[df['AI'] == 1]
    .sample(3, random_state=42)
    .to_string()
)
```

```python
# Count total trademarks per owner
total_tm = df.groupby(['owner_id', 'owner_name']).size().reset_index(name='total_tm')

# Count AI-related trademarks per owner
ai_tm = (
    df[df['AI'] == 1]
    .groupby(['owner_id', 'owner_name'])
    .size()
    .reset_index(name='ai_tm')
)

# Merge the two dataframes
owner_stats = total_tm.merge(
    ai_tm,
    on=['owner_id', 'owner_name'],
    how='left'
)

# Replace missing AI counts with 0
owner_stats['ai_tm'] = owner_stats['ai_tm'].fillna(0).astype(int)

# Sort by AI trademark frequency
owner_stats = owner_stats.sort_values(by='ai_tm', ascending=False)

# Print the top 5 players
print(owner_stats.head(5).to_string(index=False))
```

```python
# Import required libraries
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import nltk
import re
import string
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity

```

```python
# Define the file path inside your Google Drive
# NOTE: update the path!
file_path = "/content/drive/MyDrive/Text Analysis Workshop Stavanger 11-12 May 2026/data/euipo_tm_data.csv"

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# To speed up, let's select a deterministic random sample of 1000 documents
df = df.sample(n=1000, random_state=42).reset_index(drop=True)

# Display the first rows
df.head()
```

```python
# Cleaning the TM descriptions as we did in the previous mini-task

# Download stopwords (only needed once)
nltk.download('stopwords')

# Create a list of English stopwords
stop_words = set(stopwords.words('english'))

# Define the text preprocessing function
def preprocess_text(text):
    """
    This function:
    1. Converts text to lowercase
    2. Removes punctuation
    3. Removes stopwords
    """

    # Check for missing values
    if pd.isna(text):
        return ""

    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(f"[{re.escape(string.punctuation)}]", " ", text)

    # Split text into words
    words = text.split()

    # Remove stopwords
    processed_words = [
        word
        for word in words
        if word not in stop_words
    ]

    # Join words back into a single string
    return " ".join(processed_words)
```

```python
# Apply the function to the full_description column
df["clean_description"] = df["full_description"].apply(preprocess_text)

# Display results
df[["full_description", "clean_description"]].head()
```

```python
# Text --> TF-IDF vectors
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Learn vocabulary and create TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(df["clean_description"])

# Print matrix dimensions
print("TF-IDF matrix shape:", tfidf_matrix.shape)
#Showing number of documents and number of unique words/features
```

```python
# Calculating document similarity using TF-IDF
from sklearn.metrics.pairwise import cosine_similarity

# Computing cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix)

# Displaying matrix dimensions
print(similarity_matrix.shape)

# rows = documents
# columns = documents
# values = similarity scores
```

```python
# Finding the top 10 most similar non-identical document pairs

# Remove self-similarity values
np.fill_diagonal(similarity_matrix, 0)

# Create a copy of the similarity matrix
sim_matrix = similarity_matrix.copy()

# Remove identical document pairs
for i in range(len(df)):
    for j in range(len(df)):
        if df['clean_description'].iloc[i] == df['clean_description'].iloc[j]:
            sim_matrix[i, j] = 0

# Store results
top_pairs = []

# Find upper triangle indices only
# (avoids duplicate pairs like A-B and B-A)
rows, cols = np.triu_indices_from(sim_matrix, k=1)

# Collect similarity scores
for i, j in zip(rows, cols):
    top_pairs.append({
        "doc1_index": i,
        "doc2_index": j,
        "similarity": sim_matrix[i, j],
        "doc1": df['clean_description'].iloc[i],
        "doc2": df['clean_description'].iloc[j]
    })

# Convert to dataframe
top_pairs_df = pd.DataFrame(top_pairs)

# Sort by similarity
top_pairs_df = top_pairs_df.sort_values(
    by="similarity",
    ascending=False
)


top10_pairs_tfidf = top_pairs_df.head(10).reset_index(drop=True)

# Display results
top10_pairs_tfidf
```

```python
# Load pretrained Word2Vec model
model = api.load("glove-wiki-gigaword-100")

print("Model loaded.")
```

```python
# Function to create document embeddings
def document_vector(text):

    words = text.split()

    # Keep only words existing in vocabulary
    # Important: if your text is highly domain-specific,
    # you may consider training a model from scratch.
    valid_words = [
        word
        for word in words
        if word in model
    ]

    # Create vectors only for valid words
    word_vectors = [
        model[word]
        for word in valid_words
    ]

    # Calculate share of recognized vocabulary
    vocab_share = len(valid_words) / len(words) if len(words) > 0 else 0

    # Return average vector and vocabulary coverage
    if len(word_vectors) > 0:
        return np.mean(word_vectors, axis=0), vocab_share

    # Return zero vector if no valid words
    return np.zeros(model.vector_size), vocab_share

# Create vectors and store vocabulary coverage
results = [
    document_vector(text)
    for text in df["clean_description"]
]

# Separate vectors and coverage values
doc_vectors = np.array([r[0] for r in results])

df["vocab_coverage"] = [r[1] for r in results]

# Print vector matrix shape
print("Document vector matrix shape:")
print(doc_vectors.shape)

# Print average vocabulary coverage
print("\nAverage vocabulary coverage:")
print(round(df["vocab_coverage"].mean() * 100, 2), "%")

# Show first rows
df[["clean_description", "vocab_coverage"]].head()
```

```python
# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(doc_vectors)

# Print matrix dimensions
print("Similarity matrix shape:")
print(similarity_matrix.shape)
```

```python
# Remove self-similarity values
np.fill_diagonal(similarity_matrix, 0) # Every document is perfectly similar to itself (1.0). We remove these values.

# Create a copy of the matrix
sim_matrix = similarity_matrix.copy()

# Remove identical descriptions
for i in range(len(df)):
    for j in range(len(df)):

        if df["clean_description"].iloc[i] == df["clean_description"].iloc[j]:
            sim_matrix[i, j] = 0
```

```python
# Store similarity results
pairs = []

# Compare each document only once
for i in range(len(sim_matrix)):
    for j in range(i + 1, len(sim_matrix)):

        pairs.append({
            "similarity": sim_matrix[i, j],
            "doc1": df["clean_description"].iloc[i],
            "doc2": df["clean_description"].iloc[j]
        })

# Convert to DataFrame
pairs_df = pd.DataFrame(pairs)

# Top 10 most similar pairs
top10_pairs = pairs_df.sort_values(
    by="similarity",
    ascending=False
).head(10).reset_index(drop=True)

# Display results
top10_pairs
```

```python
# For curiosity sake, creating table of top 10 least similar document pairs

# Store similarity results
pairs = []

# Compare each document only once
for i in range(len(sim_matrix)):
    for j in range(i + 1, len(sim_matrix)):

        pairs.append({
            "similarity": sim_matrix[i, j],
            "doc1": df["clean_description"].iloc[i],
            "doc2": df["clean_description"].iloc[j]
        })

# Convert to DataFrame
pairs_df = pd.DataFrame(pairs)

# Top 10 least similar pairs
bottom10_pairs = pairs_df.sort_values(
    by="similarity",
    ascending=True
).head(10).reset_index(drop=True)

# Display results
bottom10_pairs
```

```python
# Output file
output_path = "/content/drive/MyDrive/Text Analysis Workshop Stavanger 11-12 May 2026/data/brightdata_sample50_classified.csv"

# Loop through dataframe
for idx, row in df.iterrows():

    firm_name = str(row["firm_name"])
    location = str(row["location"])
    website = "" if pd.isna(row["website"]) else str(row["website"])

    prompt = f"""
Describe the main products or services of {firm_name} in {location}.
Use information from the firm's website: {website} if available.

Instructions:
- Output must be in English.
- Maximum 12 words.
- Description only.
- No explanation.
"""

    try:
        response = client.responses.create(
            model="gpt-5-nano",
            input=prompt
        )

        df.at[idx, "genai_gs"] = response.output_text.strip()
        time.sleep(0.5) # To avoid API rate limits!

    except Exception as e:
        print(f"Error at index {idx}: {e}")
        df.at[idx, "genai_gs"] = None

    # Save every 10 rows
    if idx % 10 == 0:
        df.to_csv(output_path, index=False)
        print(f"Processed row {idx}")

print("Finished. File saved.")
```

### Validation Strategies

Validation is an important step when using GenAI to approximate firms’ products and economic activities. In practice, several complementary validation strategies can be used to assess the reliability and consistency of the generated outputs.


*1. Manual validation using representative samples*

One approach is to select a representative sample from the dataset and manually compare the GenAI-generated descriptions with information available on firms’ websites, LinkedIn pages, or other social media platforms. This helps assess whether the generated descriptions are consistent with firms’ self-reported products and services.


*2. Similarity-based validation using random pairs*

Another approach is to create random firm-description pairs and calculate text similarity scores. The similarity between GenAI-generated descriptions and the original firm product descriptions can then be compared with the similarity between firms and randomly assigned product descriptions. If the GenAI-generated descriptions consistently show higher similarity to the correct firms than to random firms, this provides evidence that the model captures meaningful product-related information. We discussed this approach in the “Use of GenAI for validation”.


*3. LLM-as-a-judge evaluation*

A third approach is to use another large language model as an evaluator. In this setup, a second model assesses the quality, relevance, or consistency of the outputs generated by the first model. This is often referred to as “LLM-as-a-judge.” The evaluating model can score whether the generated descriptions accurately reflect firms’ activities or whether important information is missing.


In practice, a combination of these approaches, together with case-specific validation strategies, can provide a stronger understanding of the extent to which researchers can rely on GenAI-generated outputs.
